# **Q 3b) Rule-Based Classifier**

```python
# Rule-based classifier for Telecom Complaints

complaints = [
    "I’ve been charged twice for the same service. This is really frustrating and I want a refund.",
    "My NBN connection has been down since last night. I work from home and can’t wait on the line forever.",
    "Why can’t I log into my app even after resetting the password three times?",
    "Every time I call, I get transferred between departments and no one can solve my billing issue.",
    "I cancelled my service last month but I’m still being charged on my credit card. This is unacceptable.",
    "The phone app keeps crashing whenever I try to view my usage details. I’ve reinstalled it multiple times but the problem continues."
]
```

```python
# Defining simple rule-based keyword mappings
def rule_based_classifier(complaint):
    complaint_lower = complaint.lower()

    if any(keyword in complaint_lower for keyword in ["charge", "billing", "refund", "payment", "credit card"]):
        return "Billing"
    elif any(keyword in complaint_lower for keyword in ["internet", "nbn", "down", "connection", "speed"]):
        return "Connectivity"
    elif any(keyword in complaint_lower for keyword in ["login", "password", "reset", "account access"]):
        return "Account Access"
    elif any(keyword in complaint_lower for keyword in ["cancel", "cancellation", "terminate", "termination"]):
        return "Cancellation"
    elif any(keyword in complaint_lower for keyword in ["app", "crash", "usage details", "reinstall"]):
        return "Technical Issue"
    else:
        return "Other"
```

```python
# Running classifier on complaints
for complaint in complaints:
    prediction = rule_based_classifier(complaint)
    print(f"Complaint: {complaint}\nPredicted Category: {prediction}\n")
```

# **Q 3b) Zeroshot classifier from Hugging Face**

```python
!pip install -q transformers

from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
labels = ["Billing", "Connectivity", "Account Access", "Technical Issue", "Cancellation", "Customer Service"]

complaints = [
    "I’ve been charged twice for the same service. This is really frustrating and I want a refund.",
    "My NBN connection has been down since last night. I work from home and can’t wait on the line forever.",
    "Why can’t I log into my app even after resetting the password three times?",
    "Every time I call, I get transferred between departments and no one can solve my billing issue.",
    "I cancelled my service last month but I’m still being charged.",
    "The phone app keeps crashing whenever I try to view my usage details."
]
```

```python
for complaint in complaints:
    result = classifier(complaint, labels)
    print(f"\nComplaint: {complaint}\nPrediction: {result['labels'][0]}")
```
